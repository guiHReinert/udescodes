## Método das Secantes

Devemos implementar a fórmula:
$$
x_{i+2}= x_{i+1}-\frac{f(x_{i+1})}{f(x_{i+1})-f(x_i)}(x_{i+1}-x_i)
$$

```python
from sympy import *
```

```python
x = symbols('x')
f = Lambda(x,x-sin(x**2)+4)
f(x)
```

```python
f(2.)
```

```python
plot(f(x), (x,-5,-4.5))
```

Determinemos a raiz próxima a -5.

```python
x0 = -5.5
x1 = -4.9
for i in range(1,7):
  x2 = x1-f(x1)*(x1-x0)/(f(x1)-f(x0))
  x0 = x1
  x1 = x2
x2
```

```python
f(x2)
```

```python
def secante(f,a,b,N):
  x0 = float(a)
  x1 = float(b)
  for i in range(1,N):
    x2 = x1-f(x1)*(x1-x0)/(f(x1)-f(x0))
    x0 = x1
    x1 =x2
  return x2
```
